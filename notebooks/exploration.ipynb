```python
import pandas as pd
import numpy as np
import country_converter as coco
import os
```

```python
pd.options.display.float_format = "{:,.2f}".format
```

```python
df = pd.read_csv("../data/raw/netflix_titles.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
null_counts = df.isna().sum()
null_counts[null_counts > 0]
```

```python
df["date_added"] = pd.to_datetime(df["date_added"].str.strip(), errors="coerce")
df.info()
```

```python
df[["director", "cast", "country"]] = df[["director", "cast", "country"]].fillna("Unknown")
```

```python
df[df["duration"].isna()]
```

```python
df["duration"] = df["duration"].fillna(df["rating"])
df.loc[df["rating"].str.contains("min", na=False), "rating"] = "Unknown"
```

```python
df = df.dropna(subset=["date_added", "rating"])
```

```python
df.info()
```

```python
df.head()
```

```python
df.head()
```

```python
df.info()
```

```python
df["temp_title"] = df["title"].str.strip().str.lower()
title_duplicates = df[df.duplicated(subset=["temp_title", "type"], keep=False)]
title_duplicates.sort_values(by="temp_title")[["title", "type", "release_year", "country"]]
```

```python
df["title"] = df["title"].str.strip().str.lower()
df = df.drop_duplicates(subset=["title", "type"], keep="first")
df.drop(columns=["temp_title"], inplace=True)
```

```python
df.duplicated(subset=["title", "type"]).sum()
```

```python
df["genre"] = df["listed_in"].str.split(",").str[0].str.strip()
df = df.drop(columns=["listed_in"])
```

```python
df["country"] = df["country"].str.split(",").str[0].str.strip()
```

```python
cc = coco.CountryConverter()

country_mapping = {
    "Soviet Union": "Russia",
    "West Germany": "Germany",
    "East Germany": "Germany"
}
df["country"] = df["country"].replace(country_mapping)

df["ISO"] = cc.pandas_convert(series=df["country"], to="ISO3", not_found=None)
```

```python
df.head(10)
```

```python
df["type"] = df['type'].astype("category")
df["rating"] = df["rating"].astype("category")
df["genre"] = df["genre"].astype("category")

df["cast"] = df["cast"].str.lower().str.strip()
df["director"] = df["director"].str.lower().str.strip()
```

```python
df.info()
```

```python
os.makedirs("../data/processed", exist_ok=True)
df.to_csv("../data/processed/netflix_titles_cleaned.csv", index=False)
```
